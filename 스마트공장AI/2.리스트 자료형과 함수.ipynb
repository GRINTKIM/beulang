## 리스트 자료형
- 리스트의 요소의 제한은 없다(사실은 있음)
- 리스트의 구조는 다음과 같다.

> 리스트명(이름은사용자가지정) = [요소1, 요소2, 요소3, ...]

```python
# 빈 리스트 만들기 1
a = []
a
```

```python
# 빈 리스트 만들기 2
a = list()
a
```

```python
# 리스트의 요소에는 정수가 들어갈 수 있다.
b = [1, 2, 3, 4]
b
```

```python
# 리스트의 요소에는 문자열이 들어갈 수 있다.
c = ['Life', 'is', 'too', 'short']
c
```

```python
# 리스트의 자료형은 다음과 같은 구조일 때 여러 행으로 작성가능하다.
c = ['Life', 'is',
     'too', 'short']
c
```

```python
# 리스트에는 요소가 정수, 문자열 혼합하여 들어갈 수 있다.
d = [1, 2, 'Life', 'is']
d
```

```python
# 리스트 안에 리스트가 들어갈 수 있다.
e = [1, 2, ['Life', 'is']]
e
```

## 리스트 관련 함수

- 문자열에서 사용딘 함수와 같이 '.(온점)'을 붙여서 여러가지 함수를 호출하여 사용할 수 있다.

 - append()
 - sort()
 - insert()
 - remove()
 - extend()

```python
a = [1, 2, 3, 4, 5]
b = [5, 6, 7, 9]
```

```python
a + b
```

```python
c = 5
```

```python
# a + c
```

```python
# 리스트에 요소 추가(append()) -> 많이 사용
# 추가되는 값을 하나씩 추가할 수 있다
a.append(c)
```

```python
a
```

```python
a.append([0, 6])
a
```

```python
a.append('a')
a
```

```python
# 리스트 정렬(sort)
a = [1, 4, 3, 2]
a.sort()
a
```

```python
a = ['a', 'c', 'b']
a.sort()
a
```

```python
a.reverse()
a
```

```python
# 리스트에서 find는 지원하지 않는다.
# a.find(3)
```

```python
# 리스트에 요소 삽입
a = [1, 2, 3]
a.insert(0, 6)    # 리스트이름.insert(위치, 들어갈 값)
a
```

```python
a.insert(1, 7)    # 리스트이름.insert(위치, 들어갈 값)
a
```

```python
del a[0]
a
```

```python
# 리스트 요소의 위치를 지정해서 삭제하는 del 명령과 다르게
# remove는 특정 요소값을 지정하여 제거가능하다.
a.remove(7)
a
```

```python
a = [1, 2, 3, 4, 1, 2, 3, 4, 4, 4, 4]
a.remove(4)   # remove는 여러 값이 있을 때 첫번 째 값만 제거
a
```

```python
# count() 특정 요소의 개수 세기
a.count(4)
```

```python
# 리스트 확장(extend)
a = [1, 2, 3]
a.extend([4, 5])
a
```
