## 로지스틱 회귀 실습

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression

cancer = load_breast_cancer()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# StandardScaler()로 평균이 0, 분산 1로 데이터 분포도 변환
scaler = StandardScaler()
data_scaled = scaler.fit_transform(cancer.data)

X_train, X_test, y_train, y_test = train_test_split(data_scaled,
                                                    cancer.target,
                                                    test_size=0.3,
                                                    random_state=0)
```

```python
from sklearn.metrics import accuracy_score

lr_clf = LogisticRegression()

lr_clf.fit(X_train, y_train)
```

```python
lr_preds = lr_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score

def get_clf_eval(y_test, pred):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    print('오차 행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}'.format(accuracy, precision, recall))
```

```python
get_clf_eval(y_test, lr_preds)
```

```python
from sklearn.model_selection import GridSearchCV

params = {'penalty': ['l2'],
          'C': [0.01, 0.1, 1, 1.5, 10]}

grid_clf = GridSearchCV(lr_clf, param_grid=params, scoring='accuracy', cv=3)
grid_clf.fit(data_scaled, cancer.target)
print('최적 하이퍼 파라미터: {}'.format(grid_clf.best_params_))
```

```python
# GridSearchCV의 refit으로 이미 학습이 된 estimator 반환
estimator = grid_clf.best_estimator_
estimator
```

```python
clf_preds = estimator.predict(X_test)
```

```python
get_clf_eval(y_test, clf_preds)
```

### SVM 분류 모형

```python
from sklearn import svm
# 모형 객체 생성(기본값 kernel='rbf' 적용)
svm_clf = svm.SVC()
# train data를 가지고 모형 학습
svm_clf.fit(X_train, y_train)
```

```python
svm_preds = svm_clf.predict(X_test)
```

```python
get_clf_eval(y_test, svm_preds)
```
