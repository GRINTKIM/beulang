```python
df = pd.read_csv('/content/cctv_in_seoul.csv', encoding='utf-8')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['2016년'])
plt.show()
```

```python
plt.plot(df[['2016년', '2015년']])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['2015년'], color='green', linestyle='dashed', marker='o', markerfacecolor='red', markersize=10)
plt.plot(df['2016년'])
plt.show()
```

```python
# 2015년을 y축으로, 2016년을 x축으로 하는 scatter 분산차트를 출력합니다.

plt.scatter(df['2016년'], df['2015년'], s=50)
plt.show()
```

```python
df1 = df.sort_values(by='소계', ascending=True)
df1['소계'].plot(kind='bar', grid=True, figsize=(10,5))
plt.show()
```
