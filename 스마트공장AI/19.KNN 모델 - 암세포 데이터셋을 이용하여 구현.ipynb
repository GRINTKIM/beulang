```python
ndf = rdf[['clump', 'cell_size', 'cell_shape', 'adhesion', 'epithlial', 'chromatin', 'normal_nucleoli', 'mitoses', 'class']]
ndf.head()
```

```python
# 속성(변수) 선택
X = ndf[list(ndf.columns)[:-1]]    # 독립 변수 X
y = ndf[list(ndf.columns)[-1]]
```

```python
X.head()
```

```python
X.shape
```

```python
y.head()
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=10)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# 모형 객체 생성(k=5로 설정)
knn = KNeighborsClassifier(n_neighbors=5)

# train data를 가지고 모형 학습
knn.fit(X_train, y_train)
```

```python
y_hat = knn.predict(X_test)
```

```python
y_hat.shape
```

```python
y_test.shape
```

```python
import numpy as np
y_test = np.array(y_test)
y_test
```

```python
import pandas as pd
df = pd.DataFrame({'y_hat': y_hat, 'y': y_test})
```

```python
df.head()
```

```python
df['차이'] = df['y_hat'] == df['y']
df.head()
```

```python
len(df)
```

```python
sum(df['차이'])
```

```python
print('accuracy: ', sum(df['차이']) / len(df))
```

```python
# 모형 성능 평가 - Confusion Matrix 계산
from sklearn import metrics
knn_matrix = metrics.confusion_matrix(y_test, y_hat)
knn_matrix
```

```python
# 모형 성능 평가 - 평가지표 계산
knn_report = metrics.classification_report(y_test, y_hat)
print(knn_report)
```
