```python
from selenium import webdriver
from bs4 import BeautifulSoup

driver = webdriver.Chrome('c:/py_data/chromedriver.exe')
driver.get("https://www.nate.com/")
```

```python
key = driver.find_element_by_css_selector('#bizRank > div')
```

```python
key.click()
```

```python
html = driver.page_source  # 크롬 드라이버가 띄어 놓은 html 문서를 가지고 와서 변수 html에 저장
html
```

```python
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
soup.find('ol')
```

```python
soup.find('ol').find_all('a')
```

```python
len(soup.find('ol').find_all('a'))
```

```python
keyword_rank_tag = soup.find('ol').find_all('a')
```

```python
keyword_rank_tag
```

```python
keyword_rank = []

for i in keyword_rank_tag:
    keyword_rank.append(i.text)
```

```python
list(set(keyword_rank))  # set을 쓴 이유는 혹시 중복이 있으면 없애라고 쓴 것임
```
