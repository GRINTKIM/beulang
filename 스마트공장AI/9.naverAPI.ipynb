```python
# 게시물 1000개 긁기
import time
from tqdm.notebook import tqdm
import requests
import pandas as pd


client_id = "X5e3M9FNRTEfJdHyHm5N"
client_secret = "iwhQw8IMIg"
keyword = input('네이버 블로그 검색어 입력: ')

headers = {"X-Naver-Client-Id": client_id, 
           "X-Naver-Client-Secret": client_secret}

base_url = 'https://openapi.naver.com/v1/search/blog?query='
start_url = '&start='
dis_url = '&display='

df = pd.DataFrame()

interval = 10

for page_num in tqdm(range(1,1000,interval)):
    url = base_url + keyword + start_url + str(page_num) + dis_url + '10'

    response_data = requests.get(url, headers=headers)
    time.sleep(0.1)

    if(response_data.status_code==200):
        tmp_df = pd.DataFrame(response_data.json()['items'])
        df = pd.concat([df, tmp_df])
    else:
        print("Error Code:" + response_data.status_code)

df.to_csv('../content/네이버openAPI'+keyword+'저장.csv')
```
