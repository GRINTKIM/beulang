#### 문자열 인덱싱과 슬라이싱
- 인덱싱(Indexing)이란 무엇인가를 "가리킨다"는 의미이고, 슬라이싱(Slicing)은 무엇인가를 "잘라낸다"는 의미

```python
# 프로그래밍 언어는 숫자를 0부터 센다.
# 문자열 인덱싱, 슬라이싱
a = "Life is too short, You need Python"
```

```python
a[8]
```

```python
a[24]
```

```python
a[-1]
```

```python
a[-5]
```

```python
# 슬라이싱
# a = "Life is too short, You need Python"

b = a[0] + a[1] + a[2] + a[3]
b
```

```python
c = a[:17]
c
```

```python
c = a[19:]
c
```

```python
c = a[:]
c
```

```python
# 문자열 슬라이싱의 예
a = "20010331Rainy"
date = a[:8]
date
```

```python
weather = a[8:]
weather
```

```python
year = a[:4]
month = a[4:6]
day = a[6:8]
weather = a[8:]

print('year =', year)
print('month =', month)
print('day =', day)
print('weather =', weather)
```

#### 문자열 포매팅
문자열에서 또 하나 알아야 할 것으로는 문자열 포매팅(formatting)이 있다. 이것을 공부하기 전에 다음과 같은 문자열을 출력하는 프로그램을 작성했다고 가정해보자.  
"현재 온도는 18도 입니다."  

<br>

%s = 문자열(String)  
%c = 문자 1개(character)  
%d = 정수(Integer)  
%f = 부동소수(floating-point)  
%o = 8진수  
%x = 16진수  
%% = Literal%(문자 % 자체)  

```python
# 숫자 대입
"나는 사과를 %d개 먹었습니다"%3
```

```python
"나는 사과를 %d개 먹었습니다"%(3+4)
```

```python
eat = 5
"나는 사과를 %d개 먹었습니다"%eat
```

```python
"나는 사과를 %s개 먹었습니다"%'다섯'
```

```python
eat = '다섯'
"나는 사과를 %s개 먹었습니다"%eat
```

```python
"나는 %s일 동안 사과 %d개 먹었습니다"%("오", 5)   # 순서대로 할당
```

```python
day = '오'
eat = 10
"나는 %s일 동안 사과 %d개 먹었습니다"%(day, eat)
```

```python
"I hacve %s apples"%3   # 문자열(%s)에 정수가 들어가는 것은 허용됨
```

```python
"rate is %s"%3.234
```

```python
# 정렬과 공백
"%10s" % "hi"   # 10칸을 할당해서 출력
```

```python
"%-10s" % "hi"
```

```python
"%-10sDP." % 'hi'
```

```python
"%0.4f" % 3.141592    # 소숫점 아래 4자리까지 출력
```

```python
"%10.4f" % 3.141592   # 소숫점 아래 4자리까지 출력 & 10칸을 할당해서 출력
```

#### format 함수를 사용한 포매팅
- 문자열의 format 함수를 사용하면 좀 더 발전된 스타일로 문자열 포맷을 지정할 수 있다.

```python
# 숫자 바로 대입하기
"I eat {0} apples".format(3)
```

```python
# 문자열 바로 대입하기
"I eat {0} apples".format("five")
```

```python
# 숫자 값을 가진 변수로 대입하기
number = 3
"I eat {0} apples".format(number)
```

```python
# 2개 이상의 값 넣기
number = 10
day = 3
"I ate {0} apples. So I was sick for {1} days.".format(number, day)
# 2개 이상의 값을 넣을 경우
# 문자열의 {0}, {1}과 같은 인덱스
# 항목이 format 함수의 입력값으로 순서에 맞게 바뀐다.
```

```python
number = 10
day = "three"
"I ate {1} apples. So I was sick for {0} days.".\   # {} 안의 숫자는 인덱스 위치 값
format(number, day)
```

```python
# 이름으로 넣기
"I ate {number} apples. So I was sick for {day} days.".format(number=10, day=3)
```

```python
# 인덱스와 이름을 혼용해서 넣기
"I ate {0} apples. So I was sick for {day} days.".format(10, day=3)
```

```python
# 왼쪽 정렬
"{0:<10}".format("hi")
```

```python
# 오른쪽 정렬
"{0:>10}".format("hi")
```

```python
# 가운데 정렬
"{0:^10}".format("hi")
```

```python
# 공백채우기
"{0:-^10}".format("hi")
```

```python
"{0:!<10}".format("hi")
```

```python
# f 문자열 포매팅
# 파이썬 3.6 버전부터는 f 문자열 포매팅 기능을 사용할 수 있음
name = '홍길동'
age = 30
f"나의 이름은 {name}입니다. 나이는 {age}입니다."
```

```python
'나의 이름은 ' + name + '입니다. 나이는 ' + str(age) + '입니다.'
```

```python
age = 30
f"나는 내년이면 {age+1}살이 된다."
```
