```python
import pandas as pd
import folium   # 지도 그릴 수 있게 하는 라이브러리
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/전국문화축제표준데이터.csv', encoding='euc-kr')    # encoding='euc-kr' -> 한글 파일
df.head()
```

```python
df.info()
```

```python
# 불필요한 컬럼 제거

del df['후원기관']
del df['전화번호']
del df['홈페이지주소']
del df['관련정보']
del df['소재지지번주소']
```

```python
df.info()
```

```python
df = df.rename(columns={df.columns[1]: '장소',
                        df.columns[2]: '시작일자',
                        df.columns[3]: '종료일자',
                        df.columns[6]: '기관',
                        df.columns[7]: '주소'})

df.head()
```

```python
df[df['시작일자']>='2022-01-01']    # 판다스에서는 날짜를 조건문을 비교 가능
```

```python
df[(df['기관'].str.contains('창원'))]   # contains() : '창원'이(가) 속해있는 것을 찾아줌
```

```python
df[(df['주소'].str.contains('창원'))]   # NaN 값이 있어서 오류 발생
```

```python
df.info()
```

```python
# axis = 0/1 or 'index' / 'columns'
# subset
df.dropna(axis=0, subset=['주소'], inplace=True)    #axis=0 이면 NaN 값이 있는 row를 지움, axis=1 이면 column을 지움
```

```python
df.info()
```

```python
df[(df['주소'].str.contains('창원'))]
```

```python
df[(df['주소'].str.contains('창원')) & (df['시작일자']>='2022-01-01')]
```

```python
df.dropna(axis=0, subset=['경도'], inplace=True)
```
